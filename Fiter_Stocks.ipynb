```python
#Optimized one
#https://filterna.com/php_api/marketdataa.php?id=XPEV
import requests
import json
import pandas as pd
import numpy as np
import time
```

```python
rows=[] #Type: <class 'list'>
stocks=['CAT','DUOL','AAPL','MSFT','AMZN','ZM','XPEV','TSLA','LI','NIO','ACN','MMM','AMD','ETSY','ASAN','INTC',
       'BB','SQ','CNI','PFE','FB','UBER','LYFT','NVDA','ORCL','CRM','DBX','CTSH','JNPR','CSCO',
       'ERIC','SHOP','QCOM','F','TD'] 
for sym in stocks: #[:3]:
    res = requests.get('https://filterna.com/php_api/marketdataa.php?id='+sym)
    x = json.dumps(res.json()) #Type:  <class 'str'>
    x = x.replace('{','{"Symbol": "'+sym+'", ')
    #print(x)
    y = json.loads(x) #Type: <class 'dict'>
    #print(res.text)
    print(y)#,type(y),type(x))
    time.sleep(1)
    rows.append(y) #Type: <class 'list'>
    #print(type(rows))
    #rows

table=pd.DataFrame.from_dict(rows, orient='columns')
table
```
